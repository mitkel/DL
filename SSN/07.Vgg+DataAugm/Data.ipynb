```python
from matplotlib import pyplot as plt
import torch
import torchvision as tv

source0 = tv.datasets.MNIST("../../MNIST", train = True, download = True)
DATA0 = source0.data.unsqueeze(1).float() / 255.
TARGET0 = source0.targets
set0 = torch.utils.data.TensorDataset(DATA0, TARGET0)
loader0 = torch.utils.data.DataLoader(set0, batch_size = 100)

for DATA, TARGET in loader0:
    GRID = tv.utils.make_grid(DATA, nrow = 10)
    plt.imshow(GRID.permute(1, 2, 0))
    print(TARGET, flush = True)
    plt.show()
    input()
```

```python
set1 = tv.datasets.MNIST("../../MNIST/", train=True, download = True,
                        transform = tv.transforms.ToTensor())

loader1 = torch.utils.data.DataLoader(set1, batch_size=100)

for DATA, target in loader1:
    GRID = tv.utils.make_grid(DATA, nrow = 10)
    plt.imshow(GRID.permute(1,2,0))
    print(target, flush = True)
    plt.show()
    input()
```
