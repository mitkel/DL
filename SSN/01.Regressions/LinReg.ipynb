# Linear regression

We consider the problem of polynomial fitting:
$$ \operatorname*{argmin}_p\vert \sum_i f_d(x_i) - y_i\vert^2 \quad \text{for} \quad f_d(x) = p_0 + p_1x + \ldots + p_dx^d $$


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
param = np.array([1., 0.5, -0.2, 0.1]) 
data1 = np.linspace(-10., 10., 101) 
design1 = np.array([data1**power for power in range(len(param))]).T 
target1 = design1 @ param
plt.plot(data1, target1)
plt.show()
```

```python
data0 = np.random.random(10)*20. - 10
design0 = np.array([data0**power for power in range(len(param))]).T 
target0 = design0 @ param  + np.random.random(10)*20. - 10

plt.plot(data1, target1)
plt.plot(data0, target0, 'o')
plt.show()
```

```python
param = np.linalg.solve(design0.T @ design0, design0.T @ target0)
value1 = design1 @ param

plt.plot(data1, target1)
plt.plot(data0, target0, 'o')
plt.plot(data1, value1)
plt.show()
```

```python
param = np.zeros(4)
for epoch in range(10000):
    value0 = design0 @ param
    loss = np.sum((value0-target0)**2)
    grad = 2. * (value0 - target0) @ design0
    param -= 10**(-6) * grad
    
print(epoch, loss)

value1 = design1 @ param

plt.plot(data1, target1)
plt.plot(data0, target0, 'o')
plt.plot(data1, value1)
plt.show()
```
